```python
# Store the column names as a list

# creating a list of column names
list_of_column_names = list(df.columns)
 
# displaying the list of column names
print('List of column names : ', list_of_column_names)
```

```python
# Choose one or two columns and convert the data to Numpy arrays

#Example 1

# Method 1: Convert the column using to_numpy()
array1 = df['median_income'].to_numpy()
print(array1)

# Method 2: Convert the column using df.values 
array2 = df['median_income'].values
print(array2)
```

```python
# Display data as a scatter plot or a line graph

import matplotlib.pyplot as plt

plt.scatter(array1, array2, s=5)
plt.show()
```
